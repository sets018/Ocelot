```python
from requests import get 
from bs4 import BeautifulSoup
from selenium import webdriver
from time import sleep
from random import randint
```

```python
def reset_lists():
    area = []
    price = []
    bussiness_type = []
    bedrooms = []
    bathrooms = []
    post_link = []
    neighborhood = []
    property_type = []
    post_title = []
    post_id = []
    city = []
    garages = []
    condition = []
    c = []
    estrato = []
```

```python
# Extracting data from estrato 1
page = 1
count = 0
path = r'C:\Users\ends0\Downloads\chromedriver'
driver = webdriver.Chrome(path)
while page != 23:
    url = f"https://www.fincaraiz.com.co/apartaestudios-apartamentos-casas/venta?ubicacion=Barranquilla&pagina={page}&estrato=1"
    print(url)
    driver.get(url)
    sleep(randint(1,15))
    html_soup = BeautifulSoup(driver.page_source, features='html.parser')
    posts = html_soup.find_all('div', class_= 'MuiGrid-root MuiGrid-item MuiGrid-grid-xs-12 MuiGrid-grid-sm-6 MuiGrid-grid-lg-4 MuiGrid-grid-xl-4')
    for post in posts:
        #Extracting card items (area, bedrooms, bathrooms, garages) in that order
        post_elements = post.find('div', class_ = "MuiGrid-root jss204 MuiGrid-container MuiGrid-item MuiGrid-grid-xs-12").text
        post_elements = post_elements.split("•", 3)
        #area
        if (len(post_elements) < 1):
            post_area = 0 
        else:
            post_area = int(re.search(r'\d+', post_elements[0]).group())
        area.append(post_area)
        #bedrooms
        if (len(post_elements) < 2):
            post_bdrms = 0 
        else:
            post_bdrms = int(re.search(r'\d+', post_elements[1]).group())
        bedrooms.append(post_bdrms)
        #bathrooms
        if (len(post_elements) < 3):
            post_bthrms = 0 
        else:
            post_bthrms = int(re.search(r'\d+', post_elements[2]).group())
        bathrooms.append(post_bthrms)
        #garages
        if (len(post_elements) < 4):
            post_prkns = 0 
        else:
            post_prkns = int(re.search(r'\d+', post_elements[3]).group())
        garages.append(post_prkns)
        #Extracting link
        post_link_ = "https://www.fincaraiz.com.co"+post.find('a', class_ = "MuiTypography-root MuiLink-root MuiLink-underlineHover jss214 jss193 MuiTypography-colorPrimary")['href']
        post_link.append(post_link_)
        #Extracting post hood
        post_hood = post_link_.split("/")[5].replace("-barranquilla", "")
        neighborhood.append(post_hood)
        #Extracting post type
        post_type = post.find('div', class_ = "jss210").text.split(" ")[0]
        property_type.append(post_type)
        #Extracting price (as an int)
        post_price = post.find('span', class_ = "MuiTypography-root jss201 MuiTypography-body1").text
        post_price = int("".join(list(re.findall(r'\d+', post_price))))
        price.append(post_price)
        #Extracting post title
        post_title_ = post.text
        post_title.append(post_title_)
        #Extracting post id
        post_id_ = "" +post.find('a', class_ = "MuiTypography-root MuiLink-root MuiLink-underlineHover jss214 jss193 MuiTypography-colorPrimary")['href'].split("/")[5]
        post_id.append(post_id_)
        #Extracting city 
        post_city = post_link_.split("/")[6]
        city.append(post_city)
        #Bussiness type 
        bs_type = 'Venta'
        bussiness_type.append(bs_type)
        #Estrato 
        estrato_ = 1
        estrato.append(estrato_)
        #condition
        post_used = post.find('div', class_ = "MuiChip-root jss220")
        if (post_used is not None):
            post_used = post_used.text
        else: 
            post_used = 'null'
        condition.append(post_used)
        count = count + 1
        c.append(count)
    page = page + 1
```

```python
df_posts_e1 = pd.DataFrame({'index' : c,'post_id' : post_id,'post_title' : post_title,'bussiness_type' : bussiness_type,'condition' : condition,
                         'estrato': estrato,'property_type' : property_type,'city' : city,'neighborhood' : neighborhood,'Area': area, 'bedrooms': bedrooms, 
                         'bathrooms': bathrooms,'garages' : garages,'post_link': post_link,'Price': price})
df_posts_e1.to_csv('posts_e1.csv', index=False, encoding='utf-8')
```
